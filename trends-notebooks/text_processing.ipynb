### Processando

```python
with open('../trends-data/processed_trends.txt', 'r') as file:
     arquivo = file.readline()

# arquivo
```

```python
trends = ast.literal_eval(arquivo)
# trends
```

#### Limpando os textos

```python
def remove_url(text):
    return re.sub(r'http\S+', '', text)

def remove_emoji(text):
    return text.encode('ascii', 'ignore').decode('ascii')

def remove_at(text):
    return re.sub(r'@\S+', '', text)

def clean_tweets(tweets):
    tweets = [remove_url(tweet) for tweet in tweets]
    tweets = [remove_emoji(tweet) for tweet in tweets]
    tweets = [remove_at(tweet) for tweet in tweets]
    return tweets
```

```python
def pre_processamento_textual(texto):
    nopont = [char for char in texto if char not in string.punctuation]
    nopont = ''.join(nopont)
    sms = [word for word in nopont.split() if not word.isnumeric()]
    sms = clean_tweets(sms)
    sms = ' '.join(sms)
    sms = filter(lambda x:x[0]!='#',sms)
    sms = filter(lambda x:x[0]!='/',sms)
    sms = filter(lambda x:x[0]!='\\',sms)
    sms = ''.join(sms)
    return sms
```

#### Processamento 1: {'trend_name' : [lista de tweets]}

```python
textos_processados = {}
for k in list(trends.keys()):
    textos_processados[k] = [pre_processamento_textual(texto[0]) for texto in trends[k]]
```
